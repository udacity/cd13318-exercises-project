```python
from openai import OpenAI
from pprint import pprint

client = OpenAI()

completion = client.chat.completions.create(
  model="gpt-4.1",
  messages=[
    {"role": "user", "content": "Give me a short description of what an embedding is"}
  ],
  logprobs=True,
  top_logprobs=5
)

pprint(completion.choices[0].message)
```
